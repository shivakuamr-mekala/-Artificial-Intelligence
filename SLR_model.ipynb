```python
## create a model
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
lr = LinearRegression()
X = df["Waist"].values.reshape(-1,1)
y = df["AT"].values.reshape(-1,1)
lr.fit(X,y)
```

```python
lr.coef_  # alpha
```

```python
lr.intercept_ # beta
```

```python
y = 3.45885939*72.60 - 215.98148796  ## this is your predictive line
```

```python
predictions = lr.predict(X)
```

```python
## evaluate the model
```

```python
r2_score(df["AT"],predictions)
```
