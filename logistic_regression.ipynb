```python
X = df.drop(["ATTORNEY"],axis=1)
y = df["ATTORNEY"]
```

```python
lr = LogisticRegression()
lr.fit(X,y)
```

```python
predictions = lr.predict(X)
pred_prob = lr.predict_proba(X)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y,predictions)
```

```python
df["predictions"] = predictions
```

```python
df
```

```python
class_0_prob = pred_prob[:,0]
```

```python
class_1_prob = pred_prob[:,1]
```

```python
df["prob_class_1"] =  class_1_prob
df["prob_class_0"] =  class_0_prob
```
